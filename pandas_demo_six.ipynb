```python
import pandas as pd

people = {
    "Firstname" : ["John", "Json", "James"],
    "Lastname" : ["Smith", "Web", "Bond"],
    "Location" : ["New York", "Munich", "London"],
}

people["Firstname"]
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.columns
```

```python
df.columns = ['First', 'Last', 'Place']
```

```python
df
```

```python
df.columns = [x for x in df.columns]
df
```

```python
df.columns = [x.lower() for x in df.columns]
df
```

```python
df.columns = [x.upper() for x in df.columns]
df
```

```python
df.rename(columns={'FIRST': 'First Name', 'LAST': 'Last Name', 'PLACE': 'Current Location'}, inplace=True) # case sensitive
df.columns
```

```python
df.rename(columns={'Current Location': 'Email'}, inplace=True) # case sensitive
df.columns
```

```python
df.loc[2] = ['Ian', 'Flemming', 'IanFleming@email.com']
df.loc[2]
```

```python
df.loc[1]
```

```python
df.at[1, 'Last Name'] = 'Bourne'
df
```
